# 多线程

```python
import _thread
import time


# 为线程定义一个函数
def print_time(thread_name, delay):
    count = 0
    while count < 5:
        time.sleep(delay)
        count += 1
        print("%s: %s" % (thread_name, time.ctime(time.time())))


# 创建两个线程
try:
    _thread.start_new_thread(print_time, (
        "Thread-1",
        1,
    ))
    _thread.start_new_thread(print_time, (
        "Thread-2",
        2,
    ))
except:
    print("Error: 无法启动线程")
```
