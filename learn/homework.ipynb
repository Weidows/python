# 统计数组问题

> [统计数字问题 解析](https://zhuanlan.zhihu.com/p/79692694)

```python
# 存储0到9数字的频数
c = [0 for i in range(10)]


def solve(n):
    msg = str(n)
    # 数字n的长度
    length = len(msg)
    # 首位数字 p
    p = int(msg[0])
    # 公式处理
    for i in range(10):
        c[i] += p * (length - 1) * int(pow(10, length - 2))
    # 最高位处理
    for i in range(p):
        c[i] += int(pow(10, length - 1))
    # 余项处理
    t = int(pow(10, length - 1))
    t = n % t
    if t == 0:
        c[p] += 1
        c[0] += length - 1
        return
    lenT = len(str(t))
    if lenT != length - 1:
        c[0] += (length - 1 - lenT) * (t + 1)
    c[p] += 1 + t
    return solve(t)


n = int(input())
solve(n)
# 去零
for i in range(0, len(str(n))):
    c[0] -= pow(10, i)
print(c)
```
